# Step 4. Discover what is the mean age per occupation


```python
# first we need to group the dataset by 'occupation' and next find the mean
meanAgePerOccupation = users[['occupation', 'age']].groupby('occupation').mean()
meanAgePerOccupation
```

# Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Let's extract gender and occupation from the dataset 
df = users[['gender', 'occupation']]

# Now, filter 'male' from gender in 'df'
males = df.loc[df['gender'] == 'M']
groupByOccu = df.groupby('occupation').count()

# Group by Occupation and count in males
maleCountByOccu = males.groupby('occupation').count()

maleRatioPerOccupation = maleCountByOccu / groupByOccu

maleRatioPerOccupation.sort_values(ascending=False, by='gender')
```

# Step 6. For each occupation, calculate the minimum and maximum ages


```python
df = users[['occupation', 'age']]

min_age = df.groupby('occupation').min()
max_age = df.groupby('occupation').max()

result = pd.concat([max_age, min_age], axis=1)
result
```

# Step 7. For each combination of occupation and sex, calculate the mean age


```python
df = users[['occupation', 'gender', 'age']]

AgeByOccupation_and_Gender = df.groupby(['occupation', 'gender']).mean('age')
AgeByOccupation_and_Gender
```

# Step 8. For each occupation present the percentage of women and men


```python

genderpercentage_by_occupation = df.groupby('occupation')['gender'].value_counts(normalize=True)*100
genderpercentage_by_occupation
```

# Step 1. Import the necessary libraries


```python
import pandas as pd
```

# Step 2. Import the dataset from the attached file wind.txt


```python
data = pd.read_csv('wind.txt', delimiter='\s+')
```

# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data.rename(columns={"Yr":"Year", "Mo":"Month", "Dy":"Day"}, inplace=True)
data
```

# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.


```python
def fixYear():
    data['Year'] += 1900 #

fixYear() # Fix year by converting it in 4-digit value
data
```

# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1)) # setting the index

data.drop(['Year','Month','Day'], axis=1, inplace=True) # dropping first 3 columns Year, Month & Day

data
```

```python
data.index # To make sure its datatype is datetime64
```

# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

```python
data.isna().sum() # Null values in each column
```

```python
data.isnull().sum().sum() # Sum of all null values i.e. count of all null values in the dataset
```

# Step 7. Compute how many non-missing values there are in total.


```python
data.notnull().sum().sum() # Count of non-missing values
```

# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset.

```python
data.mean().mean() # mean of windspeeds of all locations and all the times in entire dataset
```

# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days. A different set of numbers for each location.

```python

stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
# Rename the columns with proper labels
stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 
stats
```

# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.

```python
# New daily_stats dataframe having 4 columns
dayStats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
# Rename the columns with proper labels
dayStats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)
dayStats
```

# Step 11. Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.

```python
avg_windspeed_in_jan = data[data.index.month == 1].mean()

avg_windspeed_in_jan
```

# Step 12. Downsample the record to a yearly frequency for each location.


```python
year = data.resample('Y').ffill()
year
```

# Step 13. Downsample the record to a monthly frequency for each location.


```python
month = data.resample('M').ffill()
month
```

# Step 14. Downsample the record to a weekly frequency for each location.


```python
weekly = data.resample('W').ffill()
weekly
```

# Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
weeklyStats = weekly[:52]

weeklyStats.shape[0] # Total 52 weeks in 1961
```

```python
weeklyStats = pd.concat([weeklyStats.mean(), weeklyStats.max(), weeklyStats.mean(), weeklyStats.std()], axis=1)
weeklyStats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)

weeklyStats
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df # let's display the dataset to make sure it's good to use.
```

```python
fig = plt.figure()
plt.rcParams["figure.figsize"] = [15, 10] # set the figure size

# Set x & y axis and legends
df.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

# Set the Title
plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color = 'black', fontsize = 22)

# Set x & y axis labels
plt.xlabel('YEAR', color = 'RED', fontsize = 18)
plt.ylabel('CAPITA', color = 'GREEN', fontsize = 18)

plt.show()
```
